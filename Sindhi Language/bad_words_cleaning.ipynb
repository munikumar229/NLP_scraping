```python
import os

# Function to check if a file contains any bad words
def contains_bad_words(text, bad_words):
    # Convert the text to lowercase for case-insensitive comparison
    text_lower = text.lower()
    # Check if any bad word is present in the text
    return any(bad_word in text_lower for bad_word in bad_words)

# Function to process the files
def process_txt_files(input_folder, output_folder, bad_words):
    # Ensure output folder exists
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # List all text files in the input folder
    for filename in os.listdir(input_folder):
        if filename.endswith(".txt"):
            input_file_path = os.path.join(input_folder, filename)

            # Read the contents of the file
            with open(input_file_path, 'r', encoding='utf-8') as file:
                content = file.read()

            # Skip the file if it contains any bad words
            if contains_bad_words(content, bad_words):
                print(f"Skipping file: {filename} (contains bad words)")
                continue

            # Write the new content to the new folder
            output_file_path = os.path.join(output_folder, filename)
            with open(output_file_path, 'w', encoding='utf-8') as new_file:
                new_file.write(content)
            print(f"Processed file: {filename}")


# Path to the folder containing text files
input_folder = "path_to_your_input_folder"
# Path to the folder where cleaned files will be saved
output_folder = "path_to_your_output_folder"

# List of bad words in Sindhi language
bad_words = ['badword1', 'badword2', 'badword3']  # Add all your bad words here

# Process the files
process_txt_files(input_folder, output_folder, bad_words)
```

```python
import os
import re

# Function to replace bad words with empty spaces
def replace_bad_words(text, bad_words):
    # Create a regex pattern for bad words, case-insensitive
    pattern = '|'.join([re.escape(word) for word in bad_words])
    # Replace bad words with empty spaces
    clean_text = re.sub(pattern, '', text, flags=re.IGNORECASE)
    return clean_text

# Function to split text into sentences (simple approach)
def split_into_sentences(text):
    # Assuming Sindhi sentences end with full stops (".") or other sentence delimiters
    sentences = re.split(r'(?<=[.؟!])\s+', text)
    return [sentence.strip() for sentence in sentences if sentence.strip()]

# Function to process a text file, remove bad words, and split into sentences
def process_text_file(input_file, bad_words, sentence_folder):
    # Ensure the sentence folder exists
    if not os.path.exists(sentence_folder):
        os.makedirs(sentence_folder)

    # Read the content of the file
    with open(input_file, 'r', encoding='utf-8') as file:
        content = file.read()

    # Split the content into sentences
    sentences = split_into_sentences(content)

    # Process each sentence and save as a separate text file
    for i, sentence in enumerate(sentences):
        cleaned_sentence = replace_bad_words(sentence, bad_words)
        sentence_file_path = os.path.join(sentence_folder, f"sentence_{i+1}.txt")
        with open(sentence_file_path, 'w', encoding='utf-8') as sentence_file:
            sentence_file.write(cleaned_sentence)
        print(f"Processed sentence {i+1}: {cleaned_sentence}")

# Function to merge all text files back into a single file
def merge_sentences_to_file(sentence_folder, output_file):
    with open(output_file, 'w', encoding='utf-8') as outfile:
        # Iterate through all files in the sentence folder
        for sentence_file in sorted(os.listdir(sentence_folder)):
            sentence_file_path = os.path.join(sentence_folder, sentence_file)
            with open(sentence_file_path, 'r', encoding='utf-8') as infile:
                outfile.write(infile.read() + '\n')
    print(f"All sentences merged into {output_file}")


    # Input text file in Sindhi
input_file = "/home/hp/Desktop/NLP/Hugging Face/corpus/ccmatrix.txt"
    
    # Folder to store individual sentence text files
sentence_folder = "/home/hp/Desktop/NLP/Hugging Face/newcorpus"
    
    # # Output file where all cleaned sentences will be merged
    # output_file = "path_to_merged_output_file.txt"

    # List of bad words in the Sindhi language (replace with actual words)
bad_words = sindhi  # Add all your bad words here

    # Step 1: Process the input text file and store each sentence in a separate file
process_text_file(input_file, bad_words, sentence_folder)
    
    # # Step 2: Merge all the sentence text files back into one cleaned output file
    # merge_sentences_to_file(sentence_folder, output_file)
```
