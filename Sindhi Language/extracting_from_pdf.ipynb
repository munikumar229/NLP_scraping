```python
import pytesseract
from pdf2image import convert_from_path
import os
import io

# Function to extract text from a PDF file without storing images explicitly
def extract_text_from_pdf(pdf_path):
    # Convert PDF pages to images
    images = convert_from_path(pdf_path, dpi=100)  # Reduce DPI to speed up conversion

    all_text = []

    # Process each image (page) in-memory
    for image in images:
        # Convert the image to a byte stream (in-memory)
        img_byte_array = io.BytesIO()
        image.save(img_byte_array, format='PNG')

        # Extract text from the image using Tesseract with Sindhi language model
        text = pytesseract.image_to_string(image, config='--psm 6 -l snd')

        # Append the extracted text for each page
        all_text.append(text)

    # Join all the text together
    full_text = '\n'.join(all_text)

    return full_text

# Function to process specific PDFs (book7.pdf to book41.pdf) and save the extracted text
def process_books_range(input_folder, output_folder, start, end):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for i in range(start, end + 1):
        pdf_filename = f'book{i}.pdf'
        pdf_path = os.path.join(input_folder, pdf_filename)
        
        # Ensure the PDF file exists before processing
        if os.path.exists(pdf_path):
            # Extract text from the PDF
            extracted_text = extract_text_from_pdf(pdf_path)
            
            # Save the extracted text to the output folder as book7.txt, book8.txt, etc.
            text_file_path = os.path.join(output_folder, f'book{i}.txt')
            with open(text_file_path, 'w', encoding='utf-8') as file:
                file.write(extracted_text)
            print(f'Saved text from {pdf_filename} to {text_file_path}')
        else:
            print(f'{pdf_filename} not found.')

# Set the path to the folder containing PDFs and the folder to save text files
pdf_folder_path = '/home/hp/Desktop/NLP/Sindhisalmat'
text_folder_path = '/home/hp/Desktop/NLP/Sindhisalmat/newtext'

# Process books from 7 to 41 and save the extracted text
process_books_range(pdf_folder_path, text_folder_path, start=1041, end=1050)

```

```python
import os

def load_texts(folder_path):
    texts = []
    file_names = []
    for filename in os.listdir(folder_path):
        if filename.endswith(".txt"):
            filepath = os.path.join(folder_path, filename)
            with open(filepath, 'r', encoding='utf-8') as file:
                texts.append(file.read())
                file_names.append(filepath)
    return texts, file_names
```

```python
def shingle(text, k=5):
    words = text.split()
    return set([' '.join(words[i:i+k]) for i in range(len(words) - k + 1)])

def jaccard_similarity(set1, set2):
    return len(set1.intersection(set2)) / len(set1.union(set2))

def deduplicate_by_jaccard(folder_path, threshold=0.8):
    texts, file_names = load_texts(folder_path)
    shingles_list = [shingle(text) for text in texts]
    
    duplicates = set()
    for i in range(len(shingles_list)):
        for j in range(i + 1, len(shingles_list)):
            similarity = jaccard_similarity(shingles_list[i], shingles_list[j])
            if similarity > threshold:
                duplicates.add(file_names[j])
    
    for duplicate in duplicates:
        os.remove(duplicate)
    
    print(f"Removed {len(duplicates)} duplicate articles based on Jaccard Similarity.")

deduplicate_by_jaccard("/home/hp/Desktop/NLP/different")
```
