```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://learn.sindhila.edu.pk/learn/%D8%A2%D8%A6%DB%8C%DB%92-%D8%B3%D9%86%D8%AF%DA%BE%DB%8C-%D8%B3%DB%8C%DA%A9%DA%BE%DB%8C%DA%BA")
```

```python
urls = []

# Wait for the table to load (modify the table selector accordingly)
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//*[@id='adminForm']/table/tbody")))
for j in range(8):
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//*[@id='adminForm']/table/tbody")))
    table = driver.find_element(By.XPATH, "//*[@id='adminForm']/table/tbody")
    rows = table.find_elements(By.TAG_NAME, 'tr')
    for i in range(len(rows)):
        cells = rows[i].find_elements(By.TAG_NAME, 'td')
        
        if cells:
            link_element = cells[0].find_elements(By.TAG_NAME, 'a')  
            url = link_element[0].get_attribute("href")
            urls.append(url)
    try:
        next_button = driver.find_element(By.XPATH, "//*[text()='Next']")  
        next_button.click()  
            
    except Exception as e:
        print(f"Could not find 'Next' button. Error: {e}")
    
    print(f"Extracted URLs from {j}th page")
        

print("All extracted URLs:", urls)
# driver.quit()
```

```python
import os
folder_name = 'learn_sindhi_lang_data'
os.makedirs(folder_name, exist_ok=True)

count = 0
for i in range(0, len(urls)):
    try:
        count += 1
        file_name = str((i//2))+'_article.txt'
        file_path = os.path.join(folder_name, file_name)
        driver.get(urls[i])
        print(driver.current_url)
        text = ''
        h = driver.find_element(By.XPATH,"//*[@id='ur']/a")
        text = h.text
        total_box = driver.find_element(By.XPATH,"//*[@id='course']/article/div/div[1]/div")
        p = total_box.find_elements(By.TAG_NAME,'p')
        # print(len(h1), h1[0].text)
        # p = driver.find_elements(By.TAG_NAME, 'p')
        print(len(p))
        # print(len(p), p[8].text)
        try:
            for j in range(0, len(p)):
                # print(p[i].text)
                text+=p[j].text

                # Write text to the file
            with open(file_path, 'w' , encoding='utf-8') as file:
                file.write(text)
            print(text)    
            print(count, len(text.encode('utf-8')))
        except Exception as e:
            print(e)
    except Exception as e:
        print(e)
        continue
```
