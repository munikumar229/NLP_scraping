```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
```

```python
driver = webdriver.Chrome()
```

```python
url = "https://www.sindhiadabiboard.org/Index.html"
driver.get(url)
```

```python
c_table = driver.find_element(By.ID, 'table97')
c_list = c_table.find_elements(By.TAG_NAME, 'a')
c_links = [i.get_attribute('href') for i in c_list]
print(c_links)
```

```python
import os
folder = 'sindhiababi'
os.makedirs(folder, exist_ok=True)
count = 0
for i in range(1, len(c_links)-1):
    print('category ' + str(i) +' of '+ str(len(c_links)))
    try:
        driver.get(c_links[i])
    except Exception as e:
        print(e)
    try:
        table = driver.find_element(By.ID, 'table97')

    except Exception as e:
        print(e)

    # from third <tr>
    try:
        books_links = []
        books_links_urls = []
        table_tr = table.find_elements(By.XPATH, './tbody/tr/td')
        for m in range(2, len(table_tr)):
            books_links.extend(table_tr[m].find_elements(By.TAG_NAME, 'a'))
        books_links_urls = [o.get_attribute('href') for o in books_links] 
        print([o.get_attribute('href') for o in books_links])
    except Exception as e:
        print(e)

    for j in books_links_urls:
        print('new book')
        page_links = []
        pages_links = []
        pages_links_ele = []
        try:
            print(j)
            driver.get(j)
            pdf_table = driver.find_element(By.ID, 'table155')
            pages_links_ele = pdf_table.find_elements(By.TAG_NAME, 'a')
            pages_links = [n.get_attribute('href') for n in pages_links_ele]
            pages_links = list(set(pages_links))
            page_links = []
        except Exception as e:
            print(e)
        for k in pages_links:
            try:
                if k.__contains__('Book_page'):
                    page_links.append(k)
            except Exception as e:
                print(e)
        print('pages = ' + str(len(page_links)))
        print(page_links)
        text = ''
        for l in page_links:
            try:
                driver.get(l)
                table1 = driver.find_element(By.ID, 'table155')
                text += table1.text
            except Exception as e:
                    print(e)
        count += 1
        file_name = 'book_'+str(count)+'.txt'

        file_path = os.path.join(folder, file_name)

        with open(file_path, 'w', encoding='utf-8') as file:
            file.write(text)
```
