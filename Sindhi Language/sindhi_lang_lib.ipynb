```python
import pytesseract
from pdf2image import convert_from_path
from PIL import Image
import io

# Function to extract text from a PDF file without storing images explicitly
def extract_text_from_pdf(pdf_path):
    # Convert PDF pages to images
    images = convert_from_path(pdf_path, dpi=150)  # Reduce DPI to speed up conversion

    all_text = []

    # Process each image (page) in-memory
    for image in images:  # Processing pages 3 and 4 as per original code
        # Convert the image to a byte stream (in-memory)
        img_byte_array = io.BytesIO()
        image.save(img_byte_array, format='PNG')

        # Extract text from the image using Tesseract with Sindhi language model
        text = pytesseract.image_to_string(image, config='--psm 6 -l snd')

        # Append the extracted text for each page
        all_text.append(text)

    # Join all the text together
    full_text = '\n'.join(all_text)

    return full_text

# Example usage
pdf_path = '/content/6446ccbe99cf7.pdf'  # Update this with the actual PDF path
extracted_text = extract_text_from_pdf(pdf_path)

print(extracted_text)
```

```python
import pytesseract
from pdf2image import convert_from_path
import io
import os

# Function to extract text from a PDF file without storing images explicitly
def extract_text_from_pdf(pdf_path):
    # Convert PDF pages to images
    images = convert_from_path(pdf_path, dpi=150)  # Reduce DPI to speed up conversion

    all_text = []

    # Process each image (page) in-memory
    for image in images:
        # Convert the image to a byte stream (in-memory)
        img_byte_array = io.BytesIO()
        image.save(img_byte_array, format='PNG')

        # Extract text from the image using Tesseract with Sindhi language model
        text = pytesseract.image_to_string(image, config='--psm 6 -l snd')

        # Append the extracted text for each page
        all_text.append(text)

    # Join all the text together
    full_text = '\n'.join(all_text)

    return full_text

# Directory containing PDF files
pdf_dir = '/content/pdfs'  # Replace with the actual directory
if not os.path.exists(pdf_dir):
    os.makedirs(pdf_dir)
    print(f"Directory '{pdf_dir}' created successfully.")
else:
    print(f"Directory '{pdf_dir}' already exists.")

# Output directory for TXT files
output_dir = '/content/txts'  # Replace with the desired output directory

# Create the output directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Process each PDF file in the directory
for filename in os.listdir(pdf_dir):
    if filename.endswith('.pdf'):
        pdf_path = os.path.join(pdf_dir, filename)
        extracted_text = extract_text_from_pdf(pdf_path)

        # Generate output file name
        txt_filename = os.path.splitext(filename)[0] + '_.txt'
        output_file_path = os.path.join(output_dir, txt_filename)

        # Save extracted text to a TXT file
        with open(output_file_path, 'w', encoding='utf-8') as file:
            file.write(extracted_text)
```
