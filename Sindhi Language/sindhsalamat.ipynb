```python
import fitz  # PyMuPDF
import os

def extract_text_from_pdf(pdf_path):
    # Open the PDF file
    pdf_document = fitz.open(pdf_path)
    text = ''
    # Extract text from each page
    for page_num in range(len(pdf_document)):
        page = pdf_document.load_page(page_num)
        text += page.get_text()
    pdf_document.close()
    return text

def save_text_to_file(text, file_path):
    with open(file_path, 'w', encoding='utf-8') as file:
        file.write(text)

def process_pdfs_and_save_text(input_folder, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for i, filename in enumerate(sorted(os.listdir(input_folder))):
        if filename.endswith('.pdf'):
            pdf_path = os.path.join(input_folder, filename)
            text = extract_text_from_pdf(pdf_path)
            text_file_path = os.path.join(output_folder, f'book{i+1}.txt')
            save_text_to_file(text, text_file_path)
            print(f'Saved {filename} to {text_file_path}')

# Set the path to the folder containing the PDFs and the folder to save text files
pdf_folder_path = '/home/hp/Desktop/NLP/Sindhisalmat'
text_folder_path = '/home/hp/Desktop/NLP/Sindhisalmat/text'
process_pdfs_and_save_text(pdf_folder_path, text_folder_path)
```

```python
import os

# Function to rename PDF files
def rename_pdfs_in_folder(folder_path):
    # Get all the files in the folder
    files = sorted([f for f in os.listdir(folder_path) if f.endswith('.pdf')])

    # Loop through the files and rename them
    for i, filename in enumerate(files):
        # Create the new file name
        new_name = f'book{i+1}.pdf'
        # Get the full file paths
        old_file = os.path.join(folder_path, filename)
        new_file = os.path.join(folder_path, new_name)

        # Rename the file
        os.rename(old_file, new_file)
        print(f'Renamed: {filename} -> {new_name}')

# Example usage:
folder_path = '/home/hp/Desktop/NLP/Sindhisalmat'  # Update this with the actual folder path
rename_pdfs_in_folder(folder_path)
```
